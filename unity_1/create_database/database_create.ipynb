# Dataset Exemplares

```python
df_exemplares.head()
```

```python
col_import = ['codigo_barras','registro_sistema']
df_exemplares = df_exemplares[col_import]
```

Juntando os dfs de livros em um só

```python
df_livros = df_books.merge(df_exemplares, left_on='registro_sistema', right_on='registro_sistema', how='inner')
```

```python
df_livros.head()
```

Livros que eu peguei

```python
filtro = df_emp[df_emp['nome_usuario'] == "CARLOS EDUARDO MEDEIROS DA SILVA"]
codigos = filtro['codigo_barras'].values

for codigo in codigos:
    i = df_livros[df_livros['codigo_barras'] == codigo]
    print(f"Livro: {i['titulo'].values} : {i['sub_titulo'].values}")
```

```python
df_emp
```

```python
df_livros
```

```python
df = df_emp.merge(df_livros,on="codigo_barras",how="inner")
```

```python
df
```

```python
df['titulo_completo'] = df['titulo'].fillna('') + df['sub_titulo'].fillna('')

# Se quiser remover espaços extras
df['titulo_completo'] = df['titulo_completo'].str.strip()
```

```python
df.drop(columns=['titulo','sub_titulo'],inplace=True)
```
